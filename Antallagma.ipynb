```python
train.drop(train[train['Number_Of_Sales'] > 55000].index, inplace=True)
```

```python
num_train = len(train)
y_train1=train['Price']
y_train2=train['Number_Of_Sales']
train=train.drop('Price',axis=1)
train=train.drop('Number_Of_Sales',axis=1)
```

```python
df_all = pd.concat([train, test])
```

```python
df_all.isnull().sum()
```

```python
df_all.fillna(df_all.median(), inplace=True)
```

```python
df_all.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
df_all['Datetime']=en.fit_transform(df_all['Datetime'])
```

```python
df_all.columns
```

```python
df_all=df_all.drop('ID',axis=1)
df_all.head(n=5)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
for features in df_all:
 df_all[features]=scaler.fit_transform(df_all[features])
```

```python
df_all.head(n=5)
```

```python
print df_all.shape
```

```python
X_train = df_all[:num_train]
X_test = df_all[num_train:]
```

```python
from sklearn.cross_validation import train_test_split
X_tr1,X_te1,y_tr1,y_te1=train_test_split(X_train,y_train1)
X_tr2,X_te2,y_tr2,y_te2=train_test_split(X_train,y_train2)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


import xgboost as xgb
xgb_params = {
    'n_trees': 1000, 
    'eta': 0.05,
    'max_depth':14,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'silent': 1
}

dtr = xgb.DMatrix(X_tr1, y_tr1)
dte = xgb.DMatrix(X_te1)

num_boost_rounds = 422
model = xgb.train(dict(xgb_params, silent=0), dtr, num_boost_round=num_boost_rounds)


pred1 = model.predict(dte)

r2_1=r2_score(y_te1,pred1)
mse1=mean_squared_error(y_te1,pred1)

print("mean square error and r2 score of SVM is {},{}".format(r2_1,mse1))
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


import xgboost as xgb
xgb_params = {
    'n_trees': 1000, 
    'eta': 0.05,
    'max_depth':14,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'silent': 1
}

dtr = xgb.DMatrix(X_train,y_train1)
dte = xgb.DMatrix(X_test)

cv_result = xgb.cv(xgb_params, 
                   dtr, 
                   nfold=5,
                   num_boost_round=1000,
                   early_stopping_rounds=50,
                   verbose_eval=10, 
                   show_stdv=False
                  )
num_boost_rounds = len(cv_result)
print(num_boost_rounds)
model = xgb.train(dict(xgb_params, silent=0), dtr, num_boost_round=num_boost_rounds)


pred1 = model.predict(dte)
```

```python
id_test = test['ID']
output1=pd.DataFrame()
output1 = pd.DataFrame({'ID': id_test, 'Price': pred1})
output1.to_csv('price_new.csv', index=False)
```

```python
result = pd.read_csv("top_super4.csv")
result['Number_Of_Sales']=result['Number_Of_Sales'].clip(1)
result['Price']=result['Price'].clip(0)
```
