```python
import numpy as np
import cv2
import Person
import os
#import imutils

import smtplib
cnt_up   = 0
cnt_down = 0
import datetime
video="rtsp://admin:admin@xvidiaoffice.selfip.com:8554/cam/realmonitor?channel=1&subtype=0"
zone1=1
zone2=1
def mail_to(zone1,zone2,y):
    print("Sending mail...")
    gmailaddress = 'storemonitor@xvidia.net'
    gmailpassword = 'Media$122001@610'
    mailto = 'suresh@xvidia.net'
    subject='People Movement----Xvidia Technologies'
    msg1 ='Movement Found. Go through Link-- '+str(video)+'\nPlease go through recording '+str(y)+'\nzone1-->zone2-- '+str(zone2)+'\nzone2-->zone1--'+str(zone1)
    print(str(msg1))
    body = '\r\n'.join(['To: %s' % mailto,
                    'From: %s' % gmailaddress,
                    'Subject: %s' % subject,
'', msg1])
    mailServer = smtplib.SMTP('smtp.gmail.com' , 587)
    mailServer.starttls()
    mailServer.login(gmailaddress , gmailpassword)
    mailServer.sendmail(gmailaddress, mailto , body)
    print(" \n Sent!")
    mailServer.quit()
    return
def strt_rec(cap,count,now,zone,zone1,zone2):
    x='zone_counting/people_zone'+str(zone)+'_'+str(count)+'.avi'
    y='zone_counting/people_zone'+str(zone)+'_'+str(count)+'.mp4'
    out = cv2.VideoWriter(str(x),cv2.VideoWriter_fourcc('M','J','P','G'), 24, (640,480))
    i=0
    print("Recording Starts....")
    while(i<30):
        ret,frame=cap.read()
        out.write(frame)
        i+=1
    tar="ffmpeg -i "+str(x)+" "+str(y)
    os.popen(str(tar))
    #os.popen("sudo rm -r -f "+str(x))
    print("Recording done to ",y)
    if zone==1:
       zone1+=1
    else: zone2+=1
    cursorObject.execute("INSERT INTO camera_zone(CameraName,CameraRtsp,CameraMp4,LogTime) VALUES ('Office1-Xvidia','"+str(video)+"','"+str(y)+"','"+str(now)+"')")
    connectionObject.commit()
    mail_to(zone1,zone2,y)
    return (zone1,zone2)
    
#cap = cv2.VideoCapture('AirtelStore2178/2017-12-18/AirtelStore2178.100903.mp4')
cap = cv2.VideoCapture(str(video))
#cap = cv2.VideoCapture("rtsp://admin:admin@192.168.1.201:554/cam/realmonitor?channel=1&subtype=0")

ret,frame=cap.read()

frame=frame[0:480,200:600]
rows,cols,_= frame.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),330,1)
frame = cv2.warpAffine(frame,M,(cols,rows))
#frame= imutils.rotate(frame, 330)

h,w,channels=frame.shape
#cap.set(5,24)
out1 = cv2.VideoWriter('in_out_o.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 24, (400,480))
out = cv2.VideoWriter('in_out_rec.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 24, (400,480))
print(w,h)
frameArea = h*w
areaTH = frameArea/65

line_up = int(3*(h/5))
line_down   = int(4*(h/5))

up_limit =   int(2*(h/5))
down_limit = int(5*(h/5))

print ("Red line y:",str(line_down))
print ("Blue line y:", str(line_up))
line_down_color = (255,0,0)
line_up_color = (0,0,255)
pt1 =  [0,line_down];
pt2 =  [w,line_down];
pts_L1 = np.array([pt1,pt2], np.int32)
pts_L1 = pts_L1.reshape((-1,1,2))
pt3 =  [0, line_up];
pt4 =  [w, line_up];
pts_L2 = np.array([pt3,pt4], np.int32)
pts_L2 = pts_L2.reshape((-1,1,2))

pt5 =  [0, up_limit];
pt6 =  [w, up_limit];
pts_L3 = np.array([pt5,pt6], np.int32)
pts_L3 = pts_L3.reshape((-1,1,2))
pt7 =  [0, down_limit];
pt8 =  [w, down_limit];
pts_L4 = np.array([pt7,pt8], np.int32)
pts_L4 = pts_L4.reshape((-1,1,2))

fgbg = cv2.createBackgroundSubtractorMOG2(detectShadows = True)


kernelOp = np.ones((3,3),np.uint8)
kernelOp2 = np.ones((5,5),np.uint8)
kernelCl = np.ones((11,11),np.uint8)

font = cv2.FONT_HERSHEY_SIMPLEX
persons = []
max_p_age = 5
pid = 1
count1=0
count2=0
while(cap.isOpened()):

    
    ret, frame = cap.read()
     
    frame=frame[0:480,200:600]
    #frame= imutils.rotate(frame, 330)
    rows,cols,_= frame.shape

    M = cv2.getRotationMatrix2D((cols/2,rows/2),330,1)
    frame = cv2.warpAffine(frame,M,(cols,rows))
    out.write(frame)

    fgmask = fgbg.apply(frame)
    fgmask2 = fgbg.apply(frame)
    #cv2.imshow('fgmask2',fgmask2)

    
    try:

        ret,imBin2 = cv2.threshold(fgmask2,200,255,cv2.THRESH_BINARY)
        #cv2.imshow('fgmask2',imBin2)

        mask2 = cv2.morphologyEx(imBin2, cv2.MORPH_OPEN, kernelOp2)

        mask2 = cv2.morphologyEx(mask2, cv2.MORPH_CLOSE, kernelCl)
        #cv2.imshow('fgmask2',mask2)
    except:
        print('EOF')
        print ('UP:',cnt_up)
        print( 'DOWN:',cnt_down)
        break
    
    
    
    _, contours0, hierarchy = cv2.findContours(mask2,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours0:
        area = cv2.contourArea(cnt)
        if area > areaTH:
          
            
            
            M = cv2.moments(cnt)
            cx = int(M['m10']/M['m00'])#rotational inertia
            cy = int(M['m01']/M['m00'])#center of mass
            x,y,w,h = cv2.boundingRect(cnt)
            #print('cx,cy',cx,cy)
            new = True
            if cy in range(up_limit,down_limit):
                #print('person',persons)
                for i in persons:
                    
                    if abs(cx-i.getX()) <= w and abs(cy-i.getY()) <= h:
                
                        new = False
                        i.updateCoords(cx,cy)  
                        if i.going_UP(line_down,line_up) == True:
                            cnt_up += 1;
                            count1+=1
                            goingup=1
                            now=datetime.datetime.now()
                            
                            
                            frame1=frame.copy()
                            cap1=cap
                            zone1,zone2=strt_rec(cap,count1,now,goingup,zone1,zone2)
                            cap=cap1
                            ret,frame=cap.read()
                            ret,frame=cap.read()
                            #ret,frame=read()
                            frame=frame1.copy()
                           
                        if i.going_DOWN(line_down,line_up) == True:
                            cnt_down += 1;
                            count2+=1
                            goingdown=2
                            now=datetime.datetime.now()
                            frame1=frame.copy()
                            cap1=cap
                            zone1,zone2=strt_rec(cap,count2,now,goingdown,zone1,zone2)
                            cap=cap1
                            ret,frame=cap.read()
                            ret,frame=cap.read()
                            #ret,frame=read()
                            frame=frame1.copy()
                           
                        break
                    if i.getState() == '1':
                        if i.getDir() == 'down' and i.getY() > down_limit:
                            i.setDone()
                        elif i.getDir() == 'up' and i.getY() < up_limit:
                            i.setDone()
                    if i.timedOut():
                       
                        index = persons.index(i)
                        persons.pop(index)
                        del i   
                if new == True:
                    p = Person.MyPerson(pid,cx,cy, max_p_age)
                    persons.append(p)
                    pid += 1     
            
            cv2.circle(frame,(cx,cy), 5, (0,0,255), -1)
            img = cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)            
  
        
    #
    str_up = 'OUT: '+ str(cnt_up)
    str_down = 'IN: '+ str(cnt_down)
    
    frame = cv2.polylines(frame,[pts_L1],False,line_down_color,thickness=2)
    frame = cv2.polylines(frame,[pts_L2],False,line_up_color,thickness=2)
    frame = cv2.polylines(frame,[pts_L3],False,(255,255,255),thickness=1)
    frame = cv2.polylines(frame,[pts_L4],False,(255,255,255),thickness=1)
    cv2.putText(frame, str_up ,(10,40),font,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame, str_up ,(10,40),font,0.5,(0,0,255),1,cv2.LINE_AA)
    cv2.putText(frame, str_down ,(10,90),font,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame, str_down ,(10,90),font,0.5,(255,0,0),1,cv2.LINE_AA)
    
    #cv2.imshow('Frame',frame)
    #print(frame.shape)
    out1.write(frame)
    #cv2.imshow('Mask',mask)    
    
    
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
#END while(cap.isOpened())
    

cap.release()
cv2.destroyAllWindows()
```
